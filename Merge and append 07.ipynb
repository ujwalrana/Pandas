```python
sales = pd.DataFrame({
    "name" : ["A","B","D"],
    "SP" : [30,70,40]
})
```

```python
cost = pd.DataFrame({
    "name" : ["A","B","C"],
    "CP" : [30,40,50]
})
```

```python
pd.merge(sales,cost, on=["name","name"], how="inner")
```

```python
pd.merge(sales,cost, on=["name","name"], how="left")
```

```python
pd.merge(sales,cost, on=["name","name"], how="right")
```

```python
pd.merge(sales,cost, on=["name","name"], how="outer")
```

```python
pd.merge(cost,sales,left_on=["CP"],right_on=["SP"])
```
